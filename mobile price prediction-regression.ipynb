<h1>spliting the data</h1>

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

<h1>Feature Engineering</h1>
<style>color:"pink"</style>

```python
#import
#from sklearn.preprocessing import MiniMaxScaler
from sklearn.preprocessing import StandardScaler
#class
```

```python
features_to_scale = ['Storage capacity', 'Camera quality', 'Battery life', 'Weight']
```

```python
#create object to the class
scaler = StandardScaler()
df[features_to_scale] = scaler.fit_transform(df[features_to_scale])
```

```python
#scaling the input values
x_train=scaler.fit_transform(x_train)
x_train
```

```python
x_test=scaler.transform(x_test)
x_test
```

<h1>MODEL BUILDING, TRAINING, TESTING AND EVALUATION</h1>
<style color:"red" </style>

```python
# model building
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
# model training 
lr.fit(x_train,y_train)
```

```python
# model testing
result=lr.predict(x_test)
result
```

```python
# decision tree regressor

from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()


# model training
dt.fit(x_train, y_train)

# model testing
dt_result=dt.predict(x_test)
dt_result


# model evalutaion for regression
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error
import numpy as np
mae=mean_absolute_error(y_test,dt_result)
mse=mean_squared_error(y_test,dt_result)
rmse = np.sqrt(mse)
# rmse=root_mean_squared_error(y_test,result)
print("MAE",mae)
print("MSE",mse)
print("RMSE",rmse)

```

```python
# random forest regressor

from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()

# model training
rf.fit(x_train, y_train)

# model testing
rf_result=rf.predict(x_test)
rf_result

# model evalutaion for regression
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error
import numpy as np
mae=mean_absolute_error(y_test,rf_result)
mse=mean_squared_error(y_test,rf_result)
rmse = np.sqrt(mse)
# rmse=root_mean_squared_error(y_test,result)
print("MAE",mae)
print("MSE",mse)
print("RMSE",rmse)

```

```python
# model score
train_score=lr.score(x_train,y_train)
test_score=lr.score(x_test,y_test)
print("LR MODEL TRAIN SCORE",train_score)
print("LR MODEL TEST SCORE",test_score)
```

```python
def predict_phone_quality(performance=8.5, storage_capacity=128, camera_quality=9.0, battery_life=24, weight=180, age=1.0):
    # Prepare feature list in the correct order
    features = [
        float(performance),
        int(storage_capacity),
        float(camera_quality),
        float(battery_life),
        float(weight),
        float(age)
    ]

    # Ensure the correct number of features
    if len(features) != 6:
        print("Error: Feature list does not have 6 elements.")
        print("Current features:", features)
        return

    # Convert to 2D array for scaler
    features_array = np.array([features])

    # Apply standard scaling
    features_scaled = scaler.transform(features_array)

    # Predict using the scaled input
    result = rf.predict(features_scaled)
    print("Raw model output:", result)
    return result[0]
```

```python
performance= float(input("Enter performance: "))
storage_capacity = int(input("Enter storage_capacity: "))
camera_quality = float(input("Enter camera quality: "))
battery_life = int(input("Enter battery_life: "))
weight = int(input("Enter weight: "))
age = float(input("Enter age: "))


# Prediction
result = predict_phone_quality(performance, storage_capacity, camera_quality, battery_life,weight, age)
print("Model prediction result:", result)
```
